#### Setting Metrics

```python
#Write function to calculate AIC, BIC
import math
def aic(model,X,y):
    n = len(X)
    p = len(X.columns)
    RSS = sum((y-model.predict(X))**2)
    AIC=2*p+n*math.log(RSS/n)
    return AIC
def bic(model,X,y):
    n = len(X)
    p = len(X.columns)
    RSS = sum((y-model.predict(X))**2)
    BIC = n*math.log(RSS/n)+p*math.log(n)
    return BIC
    
    
def aic_bic(model,X,y):
    AIC = aic(model,X,y)
    #n + n*np.log(2*math.pi) + n*np.log(RSS/n) + 2*(p + 1)
    BIC = bic(model,X,y)
    #n + n*np.log(2*math.pi) + n*np.log(RSS/n) + (np.log(n))*(p + 1)
    return AIC, BIC
```

```python
# Write function to calculate adjusted R-squared score:
from sklearn.metrics import r2_score
def adj_r2_score(model,X,y):
    score=r2_score(y,model.predict(X))
    adj =1-((1-score)*(len(X)-1))/(len(X)-len(X.columns)-1)
    return adj
```

Create function for model comparison

```python
#Write a function using KFold to compare the performance of every model.
from sklearn.cross_validation import KFold


def metric_score_cv(X_train,y_train,model,fold):
    kfold = KFold(len(X_train), n_folds=fold, random_state=55)
    AIC=[]
    BIC=[]
    AdjR2=[]

    for train_index, test_index in kfold:
        model.fit(X_train.iloc[train_index],y_train.iloc[train_index])
        aic, bic = aic_bic(model,X_train.iloc[test_index], y_train.iloc[test_index])
        adjr2score=adj_r2_score(model,X_train.iloc[test_index], y_train.iloc[test_index])
        AIC.append(aic)
        BIC.append(bic)
        AdjR2.append(adjr2score)
    print "AIC:{}".format(np.mean(AIC))
    print "BIC:{}".format(np.mean(BIC))
    print "Adjusted R2:{}".format(np.mean(AdjR2))
```
